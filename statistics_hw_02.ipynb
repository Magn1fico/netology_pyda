* PTRATIO: This is the pupil-teacher ratio by town
* B: This is calculated as 1000(Bk — 0.63)², where Bk is the proportion of people of African American descent by town
* LSTAT: This is the percentage lower status of the population
* MEDV: This is the median value of owner-occupied homes in 1000 dollars

Можем сделать вывод, что уровень счасться (Ladder) сильно коррелирует с уровнем социальной поддержки, ВВП на душу населения и продолжительностью жизни. Что звучит логично.

**Регрессионный анализ** - можно ли провести линейную зависимость между двумя выборками. Возьмем Ladder - удовлетворение жизнью, и generosity - доброта. Мы хотим найти прямую, которая разделит точки в пространстве наполовину. Это можно сделать графически. Нарисовать график. Можно ли сказать, что данная прямая является собственным вектором? При проецировании данных на эту прямую разброс данных максимален?

```python
generosity = data_clean1['Generosity']
ladder =  data_clean1['Ladder']
sns.regplot(x= ladder, y=generosity,data=data_clean1, fit_reg=True)
plt.xlabel('Ladder')
plt.ylabel('Generosity')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

```python
'''Построенная линия не является собтенным вектором. Она является визуальным отражением корреляции двух величин и получается с помощью метода наименьших квдратов. Собственный вектор показывает направление максимального разброса значений. Можно сказать, что у них будет схожее направление, но это разные по смыслу понятия'''
```
