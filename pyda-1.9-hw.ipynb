## Задание 1
Напишите функцию date_range, которая возвращает список дней между датами start_date и end_date. Даты должны вводиться в формате YYYY-MM-DD.

```python
import datetime
from datetime import datetime
from datetime import timedelta
```

```python
def date_range(start_date, end_date):
    start_date = datetime.strptime(start_date, '%Y-%m-%d')
    end_date = datetime.strptime(end_date, '%Y-%m-%d')
    date_list = []
    while start_date < end_date:
        start_date += timedelta(days=1)
        if start_date == end_date:
            break
        date_list.append(datetime.strftime(start_date, '%Y-%m-%d'))
    return date_list
```

```python
date_range('2019-08-21','2019-09-09')
```

## Задание 3
Дан поток дат в формате YYYY-MM-DD, в которых встречаются некорректные значения:
stream = ['2018-04-02', '2018-02-29', '2018-19-02']

Напишите функцию, которая проверяет эти даты на корректность. Т. е. для каждой даты возвращает True (дата корректна) или False (некорректная дата). 

```python
def check_date(array):
    res = []
    for idate in array:
        try:
            idate = datetime.strptime(idate, '%Y-%m-%d')
            res.append(True)
        except:
            res.append(False)
    return res
```

```python
stream = ['2018-04-02', '2018-02-29', '2018-19-02']
check_date(stream)
```

## Задание 4
В последнем примере поиска по словарю мы использовали 3 столбца. Напишите функцию, которая формирует словарь для поиска по n столбцам.

```python
def create_dic(file_name, n):
    stats_dict = {}
    with open(file_name) as f:
        for line in f:
            line = line.strip().split(',')
            stats_dict[tuple(line[:n])] = line[0]
    return stats_dict
```

```python
create_dic('Python_9_Оптимизация_вычислений_в_Python_и_большие_данные/stats.csv', 4)
```
